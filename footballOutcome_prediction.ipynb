# DISCIPLINE TREND

```python
import xml.etree.ElementTree as ET

def try_parse_int(s, default=None):
    """Tries to parse a string as an integer, returning a default value if it fails."""
    try:
        return int(s)
    except ValueError:
        return default

def parse_cards(xml_str):
    if not xml_str:
        return None  # Handle None values
    root = ET.fromstring(xml_str)
    cards_info = []
    for card in root.findall('value'):
        card_type_element = card.find('card_type')
        elapsed_element = card.find('elapsed')
        player1_element = card.find('player1')
        team_element = card.find('team')

        # Check if each element was found before accessing the 'text' attribute
        card_type = card_type_element.text if card_type_element is not None else None
        elapsed_time = try_parse_int(elapsed_element.text) if elapsed_element is not None else None
        player_id = try_parse_int(player1_element.text) if player1_element is not None else None
        team_id = try_parse_int(team_element.text) if team_element is not None else None

        # Only add card information if all elements were found
        if None not in (card_type, elapsed_time, player_id, team_id):
            cards_info.append((card_type, elapsed_time, player_id, team_id))
    return cards_info

# Apply the function to the 'card' column
df['parsed_cards'] = df['card'].apply(parse_cards)
```

```python
# Initialize an empty set to store unique values
unique_values = set()

# Define a function to update the set of unique values
def update_unique_values(values):
    if values is not None:
        unique_values.update(values)

# Apply this function to each row of the 'parsed_cards' column
df['parsed_cards'].apply(update_unique_values)

# Convert the set of unique values to a list and display it
unique_values_list = list(unique_values)
print(unique_values_list)
```

Card Type (card_type): 'y' for yellow cards.
Elapsed Time (elapsed_time): The minute in the match when the card was issued.
Player ID (player_id): The unique identifier of the player to whom the card was issued.
Team ID (team_id): The unique identifier of the player's team to whom the card was issued.

```python
# Define functions to count yellow and red cards
def count_yellow_cards(cards_info, team_id):
    if not cards_info:
        return 0  # handle None values
    return sum(1 for card in cards_info if card[0] == 'y' and card[3] == team_id)

def count_red_cards(cards_info, team_id):
    if not cards_info:
        return 0  # handle None values
    return sum(1 for card in cards_info if card[0] == 'r' and card[3] == team_id)

# Use these functions to populate the yellow and red cards columns
df['home_team_yellow_cards'] = df.apply(lambda row: count_yellow_cards(row['parsed_cards'], row['home_team_api_id']), axis=1)
df['home_team_red_cards'] = df.apply(lambda row: count_red_cards(row['parsed_cards'], row['home_team_api_id']), axis=1)
df['away_team_yellow_cards'] = df.apply(lambda row: count_yellow_cards(row['parsed_cards'], row['away_team_api_id']), axis=1)
df['away_team_red_cards'] = df.apply(lambda row: count_red_cards(row['parsed_cards'], row['away_team_api_id']), axis=1)
```

```python
# Define a function to count fouls (assuming a foul is represented by a card of any type)
def count_fouls(cards_info, team_id):
    if not cards_info:
        return 0  # handle None values
    return sum(1 for card in cards_info if card[3] == team_id)

# Use these functions to populate the fouls, yellow and red cards columns
df['home_team_fouls'] = df.apply(lambda row: count_fouls(row['parsed_cards'], row['home_team_api_id']), axis=1)
df['away_team_fouls'] = df.apply(lambda row: count_fouls(row['parsed_cards'], row['away_team_api_id']), axis=1)


# Step 1: Create discipline score for each match for each team
# Calculating discipline scores
df['home_discipline_score'] = df['home_team_fouls'] + \
    df['home_team_yellow_cards'] + 2 * df['home_team_red_cards']

df['away_discipline_score'] = df['away_team_fouls'] + \
    df['away_team_yellow_cards'] + 2 * df['away_team_red_cards']

# Step 2: Calculate the moving average to capture the trend
# Use expanding mean to include all available records up to each point

# For home teams
df['home_discipline_moving_avg'] = df.groupby('home_team_api_id')['home_discipline_score'].transform(
    lambda x: x.expanding().mean())

# For away teams
df['away_discipline_moving_avg'] = df.groupby('away_team_api_id')['away_discipline_score'].transform(
    lambda x: x.expanding().mean())

# Show some rows to verify
df[['home_team_api_id', 'home_discipline_score', 'home_discipline_moving_avg',
    'away_team_api_id', 'away_discipline_score', 'away_discipline_moving_avg']].head(10)
```

# Strength of the Opposition

```python
# Step 1: Calculate strength of opposition for each match for each team

# Calculate average goal difference for each team across all matches
average_goal_diff = pd.concat([
    df.groupby('home_team_api_id')['home_goal_diff'].mean(),
    df.groupby('away_team_api_id')['away_goal_diff'].mean()
]).groupby(level=0).mean().reset_index()
average_goal_diff.columns = ['team_api_id', 'avg_goal_diff']

# Merge this information back to the match DataFrame to get the strength of opposition
df = df.merge(average_goal_diff, how='left', left_on='away_team_api_id', right_on='team_api_id')
df.rename(columns={'avg_goal_diff': 'strength_of_opposition_home_team'}, inplace=True)

df = df.merge(average_goal_diff, how='left', left_on='home_team_api_id', right_on='team_api_id')
df.rename(columns={'avg_goal_diff': 'strength_of_opposition_away_team'}, inplace=True)

# Drop the extra columns
df.drop(columns=['team_api_id_x', 'team_api_id_y'], inplace=True)

# Show some rows to verify
df[['home_team_api_id', 'strength_of_opposition_home_team', 'away_team_api_id', 'strength_of_opposition_away_team']].head(10)
```

# Data Cleaning

```python
# Calculate the percentage of NaN in each column
nan_percentage = df.isna().mean().round(4) * 100

# Initialize a list to store the columns to delete
cols_to_drop = []

# Loop through each column and apply logic to fill or delete
for col, nan_percent in nan_percentage.items():
    if nan_percent > 0:  # Only consider columns with NaN
        if nan_percent < 10:  # If less than 10% of the data is NaN, fill with the mean
            mean_value = df[col].mean()
            df[col].fillna(mean_value, inplace=True)
        else:  # If 10% or more of the data is NaN, remove the column
            cols_to_drop.append(col)

# delete those columns
df.drop(columns=cols_to_drop, inplace=True)

# Show the resulting DataFrame to verify
df, cols_to_drop
```

```python
# List of columns to delete
columns_to_drop = [
    'id', 'country_id', 'league_id', 'match_api_id', 'home_team_api_id',
    'away_team_api_id', 'season', 'date', 'BWH', 'IWH', 'LBH', 'WHH',
    'VCH', 'BWD', 'IWD', 'LBD', 'WHD', 'VCD', 'BWA', 'IWA', 'LBA', 'WHA', 'VCA'
]

# Adding patterns for 'home_player_X' and 'away_player_X' columns
columns_to_drop.extend([col for col in df.columns if 'home_player_' in col or 'away_player_' in col or 'away_player_' in col or 'home_player_' in col])

# Check if the columns exist in the DataFrame and only keep those that do
columns_to_drop = [col for col in columns_to_drop if col in df.columns]

# Deleting the columns from the DataFrame
df_cleaned = df.drop(columns=columns_to_drop)

# Show the resulting DataFrame to verify
df_cleaned
```

# Applying XGBoost

```python
from sklearn.model_selection import train_test_split, GridSearchCV
from xgboost import XGBRegressor
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

# Step 1: Data Preparation

# Step 2: Feature Selection
targets = ['home_team_goal', 'away_team_goal']

# Step 3: Data Split
X = df_one_hot_encoded.drop(columns=targets)
y = df_one_hot_encoded[targets]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Step 4: Model Building with Hyperparameter Tuning (optimized)
param_grid = {
    'n_estimators': [200, 300],
    'learning_rate': [ 0.3,0.1],
    'max_depth': [3,5],
}

model_home = GridSearchCV(XGBRegressor(objective='reg:squarederror'), param_grid, cv=5, scoring='neg_mean_squared_error')
model_away = GridSearchCV(XGBRegressor(objective='reg:squarederror'), param_grid, cv=5, scoring='neg_mean_squared_error')

model_home.fit(X_train, y_train['home_team_goal'])
model_away.fit(X_train, y_train['away_team_goal'])

# Step 5: Model Evaluation
y_pred_home = model_home.predict(X_test)
y_pred_away = model_away.predict(X_test)



# Evaluation metrics with rounded predictions
mse_home = mean_squared_error(y_test['home_team_goal'], y_pred_home)
mse_away = mean_squared_error(y_test['away_team_goal'], y_pred_away)

r2_home = r2_score(y_test['home_team_goal'], y_pred_home)
r2_away = r2_score(y_test['away_team_goal'], y_pred_away)

best_params_home = model_home.best_params_
best_params_away = model_away.best_params_

print(f"XGBoost - Home Team:")
print(f"MSE: {mse_home}, R2: {r2_home}, Best Params: {best_params_home}")
print(f"XGBoost - Away Team:")
print(f"MSE: {mse_away}, R2: {r2_away}, Best Params: {best_params_away}")
```

```python
# Round the predictions
y_pred_home_rounded = np.round(y_pred_home).astype(int)
y_pred_away_rounded = np.round(y_pred_away).astype(int)

# Determine the winners
winners = np.where(y_pred_home_rounded > y_pred_away_rounded, 'Home', np.where(y_pred_home_rounded < y_pred_away_rounded, 'Away', 'Draw'))

print(winners)
```
